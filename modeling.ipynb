```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

import datetime
```

# Combine weather and emissions datasets

```python
df_co2 = pd.read_csv("data/DE_2023_hourly.csv")
df_co2.loc[:, 'time_unix'] = pd.to_datetime(
    df_co2['Datetime (UTC)']).dt.tz_localize('UTC').astype(int) // 10**9
df_co2.set_index("time_unix", inplace=True)
df_co2.head()
```

```python
df_weather = pd.read_csv("data/Weather_data.csv")
df_weather.set_index('time_unix', inplace=True)
df_weather.drop(columns=["Unnamed: 0"], inplace=True)
df_weather.head()
```

```python
df = df_co2.join(df_weather)
df.loc[:, "time_unix"] = df.index
df.to_csv("data/combined_data.csv")
df.head()
```

```python
# Extract weather features
sunshine_features = [key for key in df if "sunshine" in key]
wind_features = [key for key in df if "wind" in key]
radiation_features = [key for key in df if "radiation" in key]

features = sunshine_features + wind_features + \
    radiation_features + ["time_unix"]
```

```python
x = df[features]
y = df["Carbon Intensity gCO₂eq/kWh (LCA)"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import xgboost as xgb
from sklearn.metrics import mean_squared_error

x_train, x_val, y_train, y_val = train_test_split(
    x, y, test_size=0.2, random_state=42)


pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('xgboost', xgb.XGBRegressor())
])


pipeline.fit(x_train, y_train)


y_pred = pipeline.predict(x_val)

mse = mean_squared_error(y_val, y_pred)
print("Mean Squared Error:", mse)
print("Mean absolute error:", mse)
```
