```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

import datetime
```

# Combine weather and emissions datasets

```python
df_co2 = pd.read_csv("data/DE_2023_hourly.csv")
df_co2.loc[:, 'time_unix'] = pd.to_datetime(
    df_co2['Datetime (UTC)']).dt.tz_localize('UTC').astype(int) // 10**9
df_co2.set_index("time_unix", inplace=True)
df_co2.head()
```

```python
df_weather = pd.read_csv("data/Weather_data.csv")
df_weather.set_index('time_unix', inplace=True)
df_weather.drop(columns=["Unnamed: 0"], inplace=True)
df_weather.head()
```

```python
df = df_co2.join(df_weather)
df.loc[:, "time_unix"] = df.index
df.loc[:, 'dt'] = pd.to_datetime(df.index, unit='s')
df.drop(columns=["Datetime (UTC)"], inplace=True)
df.loc[:, 'month'] = df['dt'].dt.month
df.loc[:, 'hour'] = df['dt'].dt.hour
df.to_csv("data/combined_data.csv")
df.head()
```

```python
df_temp = df[df["dt"] < "2023-02-01 00:00:00"]

fig = px.line(df_temp, x="dt", y="Carbon Intensity gCO₂eq/kWh (LCA)")

fig.update_layout(
    xaxis_title="Time",
    yaxis_title="Carbon Intensity gCO₂/kWh",
    title="Carbon intensity of German electricity generation (Januar 2023)",
    width=600,
    height=300
)

fig.show()
# fig.write_image("data/carbon_intensity_plot_Jan_2023.png")
```

```python
# Extract weather features
sunshine_features = [key for key in df if "sunshine" in key]
wind_features = [key for key in df if "wind" in key]
radiation_features = [key for key in df if "radiation" in key]
```

```python
target = "Carbon Intensity gCO₂eq/kWh (LCA)"
features = wind_features + ["month", "hour"]

# random train_test split
# x = df[features]
# y = df[target]

# x_train, x_val, y_train, y_val = train_test_split(
#     x, y, test_size=0.2, random_state=42)

# time-series split
df_train = df[df["dt"] < "2023-12-01 00:00:00"]
df_val = df[df["dt"] >= "2023-12-01 00:00:00"]

x_train, y_train = df_train[features], df_train[target]
x_val, y_val = df_val[features], df_val[target]
```

```python
import time
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

model = xgb.XGBRegressor()

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('xgboost', model)
])


pipeline.fit(x_train, y_train)

y_pred = pipeline.predict(x_val)

mse = mean_squared_error(y_val, y_pred)
mape = mean_absolute_percentage_error(y_val, y_pred)

print("Mean Squared Error:", mse)
print("Mean absolute error:", mape)

unix_timestamp = str(int(time.time()))

joblib.dump(pipeline, f'models/xgb_{unix_timestamp}.pkl')
```
